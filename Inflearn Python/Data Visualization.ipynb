```python
plt.plot(y.cumsum())
plt.grid(True)       #그리드
plt.axis('tight')    #축범위
```

### 1개 데이터 = > 하나의 열 벡터(내부에 여러 값 존재)
### 여러개 데이터 => 다차원의 행렬

#### Vector = Column
#### many Vector = Matrix (Being Row)

```python
plt.plot(y.cumsum())
plt.grid(True)
plt.xlim(-1, 20) #x범위 지정
plt.ylim(np.min(y.cumsum()) - 1,
         np.max(y.cumsum()) + 1,)
```

```python
plt.figure(figsize=(8,4))

plt.plot(y.cumsum(), 'r', lw = 2, label = 'Stock Index')   #color = blue, lw = Line width
plt.plot(y.cumsum(), 'ro')



plt.grid(True)
plt.axis('tight')
plt.xlabel('Date')
plt.ylabel('Index')
plt.legend()
plt.title('Kodex200 Stock price')
```

## 2차원 data set

```python
np.random.seed(2000)
y = np.random.standard_normal((20,2)).cumsum(axis = 0)
y.shape            #20행 2열
```

```python
y
```

```python
plt.figure(figsize=(8,4))

plt.plot(y, lw = 2)
plt.plot(y, 'ro')

plt.grid(True)
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value')
plt.title('Simple Plot')
```

```python
#범례 추가

plt.figure(figsize=(8,5))

plt.plot(y[:,0], lw = 2, label = '1st')   #20행 2열에서 첫번째 열 data set = 1st
plt.plot(y[:,1], lw = 2, label = '2st')   #20행 2열에서 두번째 열 data set = 2st
plt.plot(y, 'ro')


plt.grid(True)
plt.legend(loc=3)   #범례는 사분면 상으로 표시
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value')
plt.title('A Simple Plot')
```

## 다른 스타일의 그래프

```python
y = np.random.standard_normal((1000,2))
```

```python
y  #1000 x 2 Matrix
```

```python
#산포도
plt.figure(figsize=(9,5))
plt.plot(y[:,0], y[:,1], 'ro')
plt.grid(True)
plt.xlabel('first')
plt.ylabel('second')
plt.title('Scatter plot')
```

```python
c = np.random.randint(0, 10, len(y))

plt.figure(figsize=(8,4))
plt.scatter(y[:,0], y[:,1], c = c, marker = 'o')
plt.colorbar()
plt.grid(True)
plt.xlabel('1st')
plt.ylabel('2nd')
plt.title('Scatter Plot')
```

```python
# 히스토그램

plt.figure(figsize=(7, 4))
plt.hist(y, label=['1st', '2nd'], bins=25)
plt.grid(True)
plt.legend(loc=0)
plt.xlabel('value')
plt.ylabel('frequency')
plt.title('Histogram')
```

```python
# 박스플롯

fig, ax = plt.subplots(figsize=(7, 4))
plt.boxplot(y)
plt.grid(True)
plt.setp(ax, xticklabels=['1st', '2nd'])
plt.xlabel('data set')
plt.ylabel('value')
plt.title('Boxplot')
```

```python
# 지수 함수의 적분 영역을 표시하고 레이텍 수식을 사용

from matplotlib.patches import Polygon
def func(x):
    return 0.5 * np.exp(x) + 1

a, b = 0.5, 1.5
x = np.linspace(0, 2)
y = func(x)

fig, ax = plt.subplots(figsize=(7, 5))
plt.plot(x, y, 'b', linewidth=2)
plt.ylim(ymin=0)

# 함수 내부의 값을 그래프로 표현, a ~ b 값.
Ix = np.linspace(a, b)
Iy = func(Ix)
verts = [(a, 0)] + list(zip(Ix, Iy)) + [(b, 0)]
poly = Polygon(verts, facecolor='0.7', edgecolor='0.5')
ax.add_patch(poly)

# lateX 수학공식
plt.text(0.5 * (a + b), 1, r"$\int_a^b f(x)\mathrm{d}x$",
         horizontalalignment='center', fontsize=20)

plt.figtext(0.9, 0.075, '$x$')
plt.figtext(0.075, 0.9, '$f(x)$')

ax.set_xticks((a, b))
ax.set_xticklabels(('$a$', '$b$'))
ax.set_yticks([func(a), func(b)])
ax.set_yticklabels(('$f(a)$', '$f(b)$'))
plt.grid(True)
```
