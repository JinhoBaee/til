```python
def is_duplicated_number(three_digit):
    # 조건에 따라 변환되어야 할 결과를 result 변수에 할당
    # Hint - Len과 Set을 써서 할 수 있음, 중복되는 값의 str 길이는 1 또는 2
    # Three digit 의 길이가 3일시 False 를 리턴하고 그렇지 않을시 True return
    if len(set(three_digit)) == 3:      #집합은 중복 x , 중복을 제한 길이가 세개라면
        result = False                  #False를 return
    else:
        result = True
    # ==================================
    return result
```

```python
def get_strikes_or_ball(user_input_number, random_number):
   
    i = 0     #초기 설정값 (i = loop)
    strike = 0
    ball = 0
    user_input_number = user_input_number.strip() #두 변수를 str 타입으로 취급, strip 으로 처리
    random_number = random_number.strip()
    
    while(i < 3):     #while 구문 2개 , elif 구문, else 하나씩 맞춰주기
        j = 0
        while(j < 3):
            if is_validated_number(user_input_number) == False:  
                break  #최초 validated 조건에 맞지 않는다면, 탈출해라
            
            elif (i == j) is True:              #strike 가 되던가
                if random_number[i] == user_input_number[j]: #set에서 1개의 항목 일치시
                    strike += 1
            else:                               #Ball 이 된다
                if random_number[i] == user_input_number[j]: #위 elif 구문에 속하지 않지만 일치시
                    ball += 1                               #position이 다르다
            j += 1
        i += 1
    result = [strike, ball]
    # ==================================
    return result
```
